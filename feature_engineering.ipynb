```python
df_faildates = df_all.filter(pl.col("failure") == 1)
```

```python
df_faildates = df_faildates.to_pandas()
```

```python
df_faildates
```

```python
def create_faildate(df):
    df_new = df.loc[:,["serial_number", "date"]]
    df_index = df_new.set_index("serial_number", drop=True)
    fail_dict = df_index.to_dict()["date"]
    return fail_dict
```

```python
dict_fail = create_faildate(df_faildates)
```

```python

df_with_faildates = df_all.with_columns(pl.col("serial_number").map_dict(dict_fail).alias("faildate"))
```
