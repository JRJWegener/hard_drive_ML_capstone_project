```python
import numpy as np
import pandas as pd
import polars as pl
from sklearn.linear_model import RidgeClassifierCV
from sklearn.pipeline import make_pipeline
from sklearn import metrics

from sktime.datasets import load_arrow_head  # univariate dataset
from sktime.datasets import load_basic_motions  # multivariate dataset
from sktime.transformations.panel.rocket import Rocket

import preprocessing.transform_data as td
```

```python
df_ST4000 = pl.read_parquet("./data/2022_ST4000.parquet")
```

```python
df_S300YQM3 = df_ST4000.filter(pl.col("serial_number")=="S300YQM3")
df_sorted3 = df_S300YQM3.sort(by='date').to_pandas()
```

```python
df_sorted1 = df_ST4000.sort(by='date').to_pandas()
```

```python
df_list = td.timeseries_batches(df_S300YQM3)
df_rocket = pd.DataFrame(columns= td.feature_list())

df_rocket = td.transform_rocket(df_list, df_rocket)
```

```python
(df_rocket).head()
```

```python
X_train = df_rocket.drop(['target'], axis=1)[0:18]
y_train = df_rocket.target[0:18]
```

```python
X_train
y_train = y_train.astype('int')
```

```python
rocket = Rocket()
rocket.fit(X_train)
X_train_transform = rocket.transform(X_train)
```

```python
X_train_transform
```

```python
classifier = RidgeClassifierCV(alphas=np.logspace(-3, 3, 10))
classifier.fit(X_train_transform, y_train)
```

```python
X_test = df_rocket.drop(['target'], axis=1)[18:]
y_test = df_rocket.target[18:]
X_test_transform = rocket.transform(X_test)
```

```python
X_test_transform
```

```python
y_pred = classifier.predict(X_test_transform)
```

```python
y_pred
y_test = y_test.astype('int')
```

```python
metrics.accuracy_score(y_pred, y_test)
```
